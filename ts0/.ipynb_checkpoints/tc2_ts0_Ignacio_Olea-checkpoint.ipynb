#### **Obtención de la Función Transferencia**
Para la obtensión de la transferencia se usa un amplificador operación ideal, haciendo que la corrientes que entran por los pines positivos y negativos sean igual que cero, al igual que la tensión en los pines debe ser la misma. De esta forma tenemos las herramientas para encontrar la transferencia con facilidad:

\begin{equation}
\begin{cases} 
I_{R1} = I_{R2}\\\\
I_{C1} = I_{R3} \\\\
V^{-} = V^{+}
\end{cases}
\end{equation}

Así mismo, para la obtención del módulo y la fase hay que pasar del dominio de Laplace al dominio de Fourier diciendo que $ s = j\omega_0 $ para luego pasarlo a la forma de Euler y dividir entre módulo y fase.

[Mostrar PDF](./tc2_ts0_oft.pdf)

```python
from IPython.display import IFrame
IFrame("./tc2_ts0_oft.pdf", width=1000, height=1000)
```

Así como resultado conseguimos:
##### Función Transferencia
\begin{equation}
\frac{V_2}{V_1} = \frac{ s - \frac{R2}{CR1R3} }{ s + \frac{1}{CR3} }
\end{equation}

##### Módulo
\begin{equation}
\left.|\frac{V_2}{V_1}| \right|_{s=j\omega} = \sqrt{ \frac{ (\frac{R2}{CR1R3})^2 +\omega^2 }{( \frac{1}{CR3} )^2 + \omega^2 } }
\end{equation}

##### Fase
\begin{equation}
\left.\angle\frac{V_2}{V_1} \right|_{s=j\omega} = \arctan{(-\frac{CR1R3}{R2}\omega)} - \arctan{(CR3\omega)}
\end{equation}

<a id="eyn"></a>
#### **Escalado y Normalización**
 Para la normalización, normalizamos respecto a la impedancia y a la frecuencia, de esta forma elegimos valores de $n_w$ y $n_z$:
\begin{equation}
\begin{cases} 
n_w = \frac{1}{CR3} \\\\
n_z = R3
\end{cases}
\end{equation}
    
##### Valores Normalizados de los Componentes:
\begin{equation}
\begin{cases} 
R1' = \frac{R1}{n_z} = \frac{R1}{R3} \\\\
R2' = \frac{R2}{n_z} = \frac{R2}{R3} \\\\
R3' = \frac{R3}{n_z} = \frac{R3}{R3} = 1 \\\\
C' = C.n_w.n_z = C.\frac{1}{CR3}.R3 = 1
\end{cases}
\end{equation}

##### Trasferencia Normalizada:
\begin{equation}
\frac{V_2}{V_1} = \frac{ s - \frac{R2}{R1} }{ s + 1 }
\end{equation}

 Se puede observar gracias a la transferencia normalizada y al gráfico de polos y ceros que $R1$ y $R2$ son los responsables de modificar la funcionabilidad del filtro de forma que su relación puede hacer que el filtro sea un Filtro Pasa Bajos, Filtro Pasa Altos o un Filtro Pasa Todo que no posee ganancia (debido a que no hay ninguna constante que la aporte).

 
##### Verificación de Transferencia con y sin normalizar con SymPy:

```python
plt.figure(1)
plt.close(1)

import matplotlib as mpl

fig_sz_x = 13
fig_sz_y = 7
fig_dpi = 80 # dpi
fig_font_size = 11

mpl.rcParams['figure.figsize'] = (fig_sz_x, fig_sz_y)
mpl.rcParams['figure.dpi'] = fig_dpi
plt.rcParams.update({'font.size':fig_font_size})

# Importo las librerias
import sympy as sp
from sympy import Eq
from IPython.display import display

# Declaro qué es variable, componente y tensiones.
s = sp.symbols('s')
R1, R2, R3, C = sp.symbols('R1 R2 R3 C')
V1, V2, Vx = sp.symbols('V1 V2 Vx')  # Vx = V+ = V-

# Agrego las ecuaciones de nodos
eq1 = (V1-Vx)/R1 + (V2-Vx)/R2
eq2 = (V1-Vx)*C*s - (Vx)/R3
sol = sp.solve([eq1, eq2], (Vx, V2))

# Armo la transferencia y la presento
H = sp.simplify(sol[V2] / V1)
display(Eq(sp.Function('T')(s), H))

# Agrego las ecuaciones de nodos con componentes normalizados
eq1_n = (V1-Vx)/R1 + (V2-Vx)/R2
eq2_n = (V1-Vx)*s - (Vx)
sol_n = sp.solve([eq1_n, eq2_n], (Vx, V2))

# Armo la transferencia y la presento
H_n = sp.simplify(sol_n[V2] / V1)
display(Eq(sp.Function('Tn')(s), H_n))
```

#### **Simulación de la Transferencia Normalizada en Python**

```python
# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

# Librería de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

# Variables y Transferencia
#           T($) = ( $ - R2/R1 ) / ( $ + 1 )

R1 = 1000 # Valores para ejemplificar el gráfico
R2 = 200

rel = R2 / R1

my_tf = TransferFunction( [1,-rel], [1,1] )

# Gráficos

#    Gráfico de Bode con la relación R1 y R2 tal que R1 >> R2:
bodePlot(my_tf, fig_id=1, filter_description = rf"$\frac{{R_2}}{{R_1}} = {rel:.2f}$" )

#    Gráfico de Bode con la relación R1 y R2 tal que R1 << R2:
rel = R1 / R2
my_tf = TransferFunction( [1,-rel], [1,1] )
bodePlot(my_tf, fig_id=2, filter_description = rf"$\frac{{R_2}}{{R_1}} = {rel:.2f}$" )

#    Gráfico de Bode con la relación R1 y R2 tal que R1 = R2:
rel = 1
my_tf = TransferFunction( [1,-rel], [1,1] )
bodePlot(my_tf, fig_id=3, filter_description = rf"$\frac{{R_2}}{{R_1}} = {rel:.2f}$" )
fig = plt.figure(3) # Se usó para evitar ver una señal de ruido de 10^(-15)dB
fig.axes[0].set_ylim(-2, 2)

#    Gráfico de polos para R1 = R2:
pzmap(my_tf, fig_id=4, filter_description = rf"$\frac{{R_2}}{{R_1}} = {rel:.2f}$") 
```

Si la relación $R1$ y $R2$ se hace más pequeña ($R1$ predomina), el Cero se acercará al origen y el filtro se vuelve un Pasa Altos (elimina las bajas frecuencias). Por otro lado, si la relación se hace mayor ($R2$ predomina), la forma del filtro comienza a variar pasando de Pasa Altos a un Pasa Bajos.  Por otro lado, si la relación es igual a uno ($R1$ igual que $R2$) es el equilibrió entre el Pasa Bajos y el Pasa Altos, quedando como resultado un Pasa Todo (que es equivalente a un cable ya que no tiene ganancia).

#### **¿Qué Tipo de Filtro es?**
 Ya se habló más arriba y el filtro presenta un comportamiento distinto dependiente de los valores de los componentes y la relación entre ellos, más especificamente, los valores de $R1$ y $R2$ por lo que no se puede decir que sea un único filtro, si no que es *a la vez* un Filtro Pasa Altos, un Filtro Pasa Bajos y un Filtro Pasa Todo. 

  Aunque no logré encontrar un fragmento de circuito dentro del MF-103 que tenga la misma red, logré encontrar uno que se le asemeja en su comportamiento:
 
 <img src="./ts0_mf-103.png" /> 

 Lo que tiene en común esta red de la dada por el docente es que cambia el comportamiento del circuito dependiendo del valor de los resistores como se muestra en el siguiente analisís: se puede observar las diferencias con la red dada, ya que en este caso la relación entre los resistores, además de ser diferente, modifica la amplitud y la fase del filtro; puede tener los polos y los ceros del lado izquierdo del eje Laplace con ciertos valores; y cuando los resistores son iguales, no se comporta como un cable (como en nuestra red) si no que se comporta como un circuito abierto (no deja pasar nada de señal). Con los valores del MF-103 esto actua como un Filtro Pasa Altos.

##### Analice similitudes y diferencias con ambas redes del TP1-Ej7:

Con la primera red se observa que es exactamente igual que la nuestra y particulariza el caso con $R1$ y $R2$ iguales por lo que nos encontramos con un Filtro Pasa Todo e inversor de fase.
Para la segunda red se observa que es más compleja, calcule la transferencia y, a continuación, use SymPy para verificarla con una normalización de $n_\omega = \frac{1}{CR}$ y $n_z = R$.
\begin{equation}
T(s) = -\frac{Rb}{Ra}.\frac{ s^2 + (2 - \frac{Ra}{Rb})s + 1 }{ s^2 + 3s +1 }
\end{equation}



```python
# Importo las librerias
import sympy as sp
from sympy import Eq
from IPython.display import display

# Declaro qué es variable, componente y tensiones.
s = sp.symbols('s')
Ra, Rb, R, C = sp.symbols('Ra Rb R C')
V1, V2, Vx = sp.symbols('V1 V2 Vx')  # Vx = V+ = V-

# Agrego las ecuaciones de nodos con componentes normalizados
eq1_7n = (V1-Vx)/Ra + (V2-Vx)/Rb
eq2_7n = (V1-Vx)/(1+1/(s)) - (Vx)*(s + 1)
sol_7n = sp.solve([eq1_7n, eq2_7n], (Vx, V2))

# Armo la transferencia y la presento
H_7n = sp.simplify(sol_7n[V2] / V1)
display(Eq(sp.Function('T')(s), H_7n))
```

Aunque el SymPy no deja mónico los polinomios, verifica mis resultados.
Acto seguido lo grafiqué con Python usando distintas resistencias para modifical el término lineal y ver como cambia el comportamiento del filtro. *NOTA: analizando el comportamiento del filtro se despreció el efecto de la constante $-\frac{Rb}{Ra}$*:

```python
# Librerías externas NumPy, SciPy y Matplotlib
from scipy.signal import TransferFunction
import matplotlib.pyplot as plt
import numpy as np

# Librería de TC2, esta la vas a usar mucho
from pytc2.sistemas_lineales import pzmap, GroupDelay, bodePlot

RA = [0,50,60] # Valores para ejemplificar el gráfico
Rb = 10

for Ra in RA:
    rel = 2- Ra / Rb

    my_tf = TransferFunction( [1,rel,1], [1,3,1] )

# -----------------Gráficos-----------------------

#    Gráfico de Bode 
    bodePlot(my_tf, fig_id=1)

#    Gráfico de polos 
    pzmap(my_tf, fig_id=2, filter_description = rf"$2-\frac{{R_B}}{{R_A}} = {rel:.2f}$") 
```

De esta forma podemos ya podemos hablar de similitudes y diferencias con nuestro circuito: Se observa un comportamiento parcialmente distinto, si en ambos circuitos variamos sus resistores correspondientes, nos encontramos que el primero que analizamos se comporta como un Filtro Pasa Todo, Pasa Bajos o Pasa Altos de primer orden; mientras que el circuito del ejercicio 7 se comporta como un Pasa Banda, Elimina Banda o Pasa Todo de segundo orden. De esta forma vemos que como similitudes:
1. Ambos se comportan como un Filtro Pasa Todo bajo determinados valores,
2. Sus Polos y Ceros, para el caso de Pasa Todo, son completamente reales y muestran simetria respecto al eje imaginario,
3. Ambos invierten la fase, solo si, para el segundo circuito, el termino lineal es menor a $-3$ y si se lo usa en la frecuencia igual a $\omega_0$.

Diferencias:
1. Tienen un orden de diferencia y el segundo usa más componentes, 
2. Fuera del Pasa Todo se comportan distinto tanto en comportamiento del módulo como en su fase (llegando a conseguir un filtro Notch cuando Ra es el doble que Rb).
